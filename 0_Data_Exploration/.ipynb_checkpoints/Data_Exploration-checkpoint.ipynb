```python
# Python libraries
import os
import sys
from pprint import pprint

# Additional modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# from mpl_toolkits.basemap import Basemap

# Configuration
SEED = 42
np.random.seed(SEED)
```

### Analysis Part 1: Standard Analysis
Questions to be answered:
1. What is are the datasets' shapes?
1. What features are given in the train and test sets?
2. What data types are they?
3. How many missing values does each column contain?


```python
train_X = pd.read_csv(os.path.join('..', 'data', 'train.csv'), index_col = 'RowId')
test_X = pd.read_csv(os.path.join('..', 'data', 'test.csv'), index_col = 'RowId')
print('Train set shape:', train_X.shape)
print('Test set shape:', test_X.shape)
print()

train_cols = list(train_X.columns)
test_cols = list(test_X.columns)
not_in_train = [x for x in train_cols if not x in test_cols]
print('{} test columns for prediction:'.format(str(len(test_cols))), ', '.join(test_cols), '\n')
print('{} train columns:'.format(str(len(train_cols))), ', '.join(train_cols), '\n')
print('{} columns in train but not in test:'.format(str(len(not_in_train))), ', '.join(not_in_train), '\n')

display(train_X.sample(10, random_state = SEED).drop([col for col in train_cols if 'Stop' in col], axis = 1))
```

The training dataset provides more percentiles for the target metrics and also an additional metric: `TimeFromFirstStop`.
What are the characteristics of each column?

```python
criteria = ['column name', 'contains NAs', 'data type', 'number of unique values', 'min', 'max', 'mean']
col_descriptions = {crit: [] for crit in criteria}
numeric_features = []

for col_name in train_X.columns.tolist():
    col = train_X[col_name]
    
    col_descriptions['column name'].append(col_name)
    
    num_nas = np.count_nonzero(pd.isna(col).values)
    col_descriptions['contains NAs'].append(num_nas)
    
    dtype = str(col.values.dtype)
    col_descriptions['data type'].append(dtype)
    
    col_descriptions['number of unique values'].append(len(col.unique()))
    
    if dtype == 'float64' or dtype == 'int64':
        col_descriptions['min'].append(np.round(np.min(col), 2))
        col_descriptions['max'].append(np.round(np.max(col), 2))
        col_descriptions['mean'].append(np.round(np.mean(col), 2))
        numeric_features.append(col_name)
    else:
        col_descriptions['min'].append(np.nan)
        col_descriptions['max'].append(np.nan)
        col_descriptions['mean'].append(np.nan)

display(pd.DataFrame(col_descriptions).set_index('column name'))
```

Here, you can see that the street names are the only columns with missing values which only make up a percent of the dataset.
* What units do the measurements of the target variables use? Presuming seconds are used looking at the means.
* What units used for `DistanceToFirstStop`? Assuming it is measured in meters.
* The `EntryHeading` and `ExitHeading` features can be used by applying one-hot encoding. Same with `City`.
* Higher percentlies of vehicles in a group of course wait longer and are forther away from the intersection.
* All the target metrics have outliers since the maximum values are often many times the mean.
* 28% of all observations are from weekends which is what one would expect: $2 / 7 = 0.2857$.

Let's take a look at the regressors' distributions:

```python
%matplotlib inline

plt.hist(train_X['Month'].values, bins = 12)
plt.ylabel('# Occured')
plt.title('Distribution of months')
plt.show()

plt.hist(train_X['Hour'].values, bins = 24)
plt.ylabel('# Occured')
plt.title('Distribution of hours')
plt.show()
```

* We only have observations from the second half of the year and a few observations from January
    * Excluding January, the distrubution is almost uniform.
* The hours are distributed as one would expect:
    * Least traffic when most people sleep between 10 PM and 5 AM, then suddenly spiking when most people go to work 8AM and then slowly pacing off from 3 PM on.
* Expect high correlation between waiting times and rush hours between 8 AM and 3 PM.

```python
# Comparison of percentile distributions - distance to intersection
def target_distr_comparison(xlims, feature):
    percentiles = ['_p20', '_p40', '_p50', '_p60', '_p80']
    colors = ['#ff0000', '#e50000', '#cc0000', '#990000', '#660000']
    
    for i in range(len(percentiles)):
        percentile = percentiles[i]
        color = colors[i]
        plt.hist(
            train_X[feature + percentile].values, 
            bins = 200,
            histtype = 'step',
            color = color
        )
    
    if not xlims is None:
        plt.xlim(xlims)
    plt.ylabel('# Occured')
    plt.title('Distribution of ' + feature)
    plt.show()
    
target_distr_comparison((0, 250), 'DistanceToFirstStop')
target_distr_comparison((30, 150), 'DistanceToFirstStop')

target_distr_comparison((0, 120), 'TotalTimeStopped')
target_distr_comparison((5, 30), 'TotalTimeStopped')
```
